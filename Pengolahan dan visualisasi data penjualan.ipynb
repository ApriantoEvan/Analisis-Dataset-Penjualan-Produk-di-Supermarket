```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Supermart Grocery Sales - Retail Analytics Dataset.csv")
data.head(10)
```

```python
data.drop(["Order ID", "Order Date", "State", "Sub Category", "Discount"], axis=1, inplace=True)
data.head(10)
```

```python
arr = np.array(data["Sales"])
print(f'Total penjualan: {int(arr.sum())}')
print(f'Rata-rata penjualan: {int(arr.mean())}')
print(f'Penjualan terbanyak: {int(arr.max())}')
print(f'Penjualan tersedikit: {int(arr.min())}')
```

```python
# Selection Sort algorithm in Python
def DescendingSort(array, size):
	
	for s in range(size):
		min_idx = s
		
		for i in range(s + 1, size):

			if array[i] > array[min_idx]:
				min_idx = i

		(array[s], array[min_idx]) = (array[min_idx], array[s])

# Driver code
datal = list(data['Sales'])
size = len(datal)
DescendingSort(datal, size)

print('Urutan penjualan dari yang terbanyak :')
dfl= pd.DataFrame(datal, columns = ['Sales'])
dfl.head(20)
```

```python
# Selection Sort algorithm in Python
def AscendingSort(array, size):
	
	for s in range(size):
		min_idx = s
		
		for i in range(s + 1, size):

			if array[i] < array[min_idx]:
				min_idx = i

		(array[s], array[min_idx]) = (array[min_idx], array[s])

# Driver code
datal = list(data['Sales'])
size = len(datal)
AscendingSort(datal, size)

print('Urutan penjualan dari yang tersedikit :')
dfl= pd.DataFrame(datal, columns = ['Sales'])
dfl.head(20)
```

```python
# Linier search algorithm in python
def search(arr, N, x):

	for i in range(0, N):
		if (arr[i] == x):
			return i
	return -1


# Driver Code
if __name__ == "__main__":
	arr = list(data['Customer Name'])
	x = input('Masukkan nama customer yang ingin dicari: ')
	N = len(arr)

	# Function call
	result = search(arr, N, x)
	if(result == -1):
		print(f"Tidak terdapat {x} dalam list")
	else:
		print(f"{x} berada dalam list di index {result}")
```

```python
s = data['Category'].value_counts()
print(s)
```

```python
e = s.reset_index(name='Count')
dfe = e.rename(columns = {'index' : 'Food Category'})
print(dfe)
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = dfe['Food Category']
sizes = dfe['Count']
explode = (0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

print('Berikut ini adalah pie chart penjualan berdasarkan kategori produk \n')
plt.show()
```
